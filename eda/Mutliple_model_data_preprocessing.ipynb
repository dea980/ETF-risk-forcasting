## 각 데이터 별, 전처리 기본적인 방법

### 정형 데이터 전처리
    -   ETF 시계열: Close, LogReturn, Volatility, Drawdown
    -   기술적 지표: RSI, MACD, Bollinger Bands
    -   생성한 전략 Signal 값

    1.기본 전처리 단계
    -   결측치 처리 (dropna 또는 fillna)

    -   정규화 / 표준화 (StandardScaler, MinMaxScaler)

    -   시계열 시퀀스 생성 (LSTM, Transformer용)

    -   타겟 생성 (예: 미래 수익률 또는 시그널)
### 비정형 데이터 전처리
    -   캔들 차트 PNG
    -   누적 수익률/변동성 시각화 이미지
    -   리포트 (RAG 질의응답용 문서)
    이미지 전처리 단계 (CNN, ViT 등)
    -   사이즈 통일 (resize → ex. 224x224)
    -   정규화 (/255 또는 EfficientNet.preprocess_input)
    -   배치 차원 추가 (np.expand_dims)
    -   (선택) Augmentation (회전, 노이즈)
